<h1><center>Comparing Distribution Of Clustered Neighbourhoods in Toronto and London</center></h1>

<h3><center>Introduction</center></h3>

Toronto and London are the most populous cities in their respective countries [1][2] but there is an obvious difference in their economic capabilities. London had a GDP of approximately \\$600BN as of 2016 [3] and Toronto has \\$300BN in 2014 [3].

A company expanding their business from Toronto to London or vice versa needs to consider many factors that will affect their growth in a new city and one of the main ways to ensure consistent growth is being able to show your prescence in a new location through advertising.
Optimising advertisement locations is a great way to reach the target audience while reducing costs.

Neighbourhoods with similar amenities and venues in close proximity tend to attract people with similar hobbies, interests or needs. This means that advertising within neighbourhoods containing your target audience is far more effective and cheaper than broadcasting across the entire city. Therefore, it is advantageous to study the distribution of similar neighbourhoods through data provided by Foursquare.

<h3><center>Data</center></h3>

The main method to segment cities into smaller neighbourhood areas was to separate by postal code. The list of all postal codes pertaining to each city can be found on Wikipedia [4][5]. Data was scraped from each page and put into a separate table for each city.

The cities of Toronto and London are very large and so it would not be practically feasible to use the limited Foursquare API calls to canvass the entirety of both cities. For Toronto, this was not a large issue as many of the postal codes within Toronto were not assigned to a borough so they were omitted. London was a tougher choice as it had about 500 postal codes and this covered a very large portion of the city. The choice made was to only focus on the centre of London where there were approximately 150 postal codes. This also provides a much closer sample number to Toronto allowing for comparisons to be more direct.

Both cities had postal codes that were readily available but to successfully compare the distribution of the neighbourhoods we also needed latitude and longitude to allow for more precise and accurate API calls and results. Boroughs within both of the cities are relatively large and so the latitude and longitude of the centre of the borough was taken. The latitude and longitude for London was found based on the centre of the postal codes and in cases where multiple postal codes were assigned to the same area, the first postal code was used.

There is also a case for multiple neighbourhoods to be assigned to a single borough and to combat this issue, the neighbourhoods were grouped into the same boroughs for both cities. London posed a further problem with multiple postcodes being assigned to the same borough. In most cases this is due to the intersection of a postcode with another borough and so the postal code listed first was taken as the main.

<h3><center>Methodology</center></h3>

To begin, we can start by looking at the top 50 venues from the location of the borough centres. This is likely to be where the most variety of venues will be but will also represent the majority of consistent venues if there are any.

After sorting through the top 50 venues in each borough for each city, we can further narrow it down by considering the top 10 most frequent venues in each borough. The top 10 most frequent venues are likely to signify the type of neighbourhood it is and therefore also the people that are likely to reside within it.

Below is the table that shows the 10 most frequent venues in 5 boroughs for London.



From the above table we can see that it would be difficult to group together similar neighbourhoods through conventional methods as there are too many categories of venues as well as the sheer quantity that is present.

We can also compare the same table for London with the top 10 most frequent venues that are seen in 5 Toronto boroughs.



A practical solution to this type of classification problem is to use k means clustering. This machine learning technique will allow us to group similar neighbourhoods together based on the frequency of the venue category and how often it appears within each neighbourhood. After applying this clustering technique, we can view how the clusters are distributed on a map and discuss how best to approach the marketing problem.

We can now cluster each city using the k means machine learning algorithm to see how the neighbourhoods are spread out.

Each city has now been described on its own and its segmentation of alike neighbourhoods can be seen but this will not completely solve our problem of where the business can advertise their products.
To do this, we will need to combine the tables from both cities and cluster them based on how alike they are without considering the geological distance.

<h3><center>Results</center></h3>

After having undergone the process of how the comparison of the two cities was carried out, lets look at how the distribution of neighbourhoods varies from city to city.

Below is the city of Toronto with the k means clustering approach applied to it. The neighbourhoods have been separated based on how alike they are to each other or dissimilar to other clusters.

As you can see there is a clear segmentation between the neighbourhoods, even the similar ones. The centre of Toronto is dominated by one kind of neighbourhood whereas the other clusters seems to be on the outskirts of the city.



Now lets compare that map of Toronto with the map of London and how they vary.



The colours do not represent the same kind of neighbourhood but we can clearly see how the distribution varies with each city. London has a very even spread of alike neighbourhoods and there isn't the clear segmentation as seen with Toronto. There are definitely clusters that are spread towards the edge of the city but it is not as prominent.

This data on its own has value within some context but it cannot be considered sufficient to dictate how the business should go about spreading their advertising budget to each area.
To allow us to do this, lets look at the outcome when we combine the data from each city and how that impacts the clustering.

This is the map of Toronto after the combining of the London and Toronto data.



Maybe surprisingly, we can still see a clear segmentation of alike neighbourhoods within the Toronto city area and while the clusters are distributed in a different manner to the stand alone data, they are still centralised to a certain area.

Now lets look at the London map within this same data.

Again, we can see that combining the data of both cities did not change the overall spread of clusters. We do see a lack of red clusters within London but it is clear that the overall distribution did not change and that there is an even spread across the city.

From this we can conclude relatively easily that advertising within London will pose a great challenge as the alike neighbourhoods do not align to a specific area and so the entire city will need to be targeted. A useful insight from this data that can be derived is if the target audience of the business resides within a particular type of neighbourhood in Toronto we can target the same neighbourhoods in London with a relatively high accuracy as we can see the similar neighbourhoods within both cities.

<h3><center>Conclusion</center></h3>

For businesses moving from Toronto to London, they need to visualise how the neighbourhoods are spread out within each city and how the new distribution of neighbourhoods will affect their advertising budget. When compared individually, we see clear segmentation in Toronto but little to no segmentation in London. This can be due to many factors such as the difference in culture or the spread of amenities between neighbourhoods but the fact that this segmentation still persists after combining the data could be an indication that the spread of similar neighbourhoods is independent of location as the services or venues is all locations are similar or easily accessible.

Overall, for businesses moving from Toronto to London will have a harder time targeting the exact type of customers they require but they will be able to easily target the similar neighbourhoods that their customers within Toronto reside in.
